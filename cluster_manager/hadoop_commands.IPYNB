#### HDFS Commands

- **hdfs dfs**: List all HDFS Commands.

- **hdfs dfs -usage ls**: Basic usage for any command.

- **hdfs dfs -help ls**: Full detailed information for any commands.

- **hdfs dfs -ls DIR_HDFS**: List dir HDFS.

- **hdfs dfs -ls -R DIR_HDFS**: List ALL Folders and files(Recursive).

- **hdfs dfs -mkdir -p /application/retail_db/**: Create DIR.

- **hdfs dfs -rm -r dir**: Remove DIR.

- **hdfs dfs -put * /DIR**: Copy ALL files to HDFS.

- **hdfs dfs -get * /DIR_HDFS DIR_LOCAL**: Copy HDFS to Local
